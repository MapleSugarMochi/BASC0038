```python
def is_prime(n: int) -> bool:
  """
  Return True if n is a prime number, False otherwise
  """
  if n < 2:
    return False
  for i in range(2, isqrt(n) + 1): # Only need to check divisors up to isqrt(n) + 1, since any factor larger than isqrt(n) + 1 would pair with a smaller one already checked
    if n % i == 0:
      return False
  return True
```

```python
# tests for is_prime()
assert is_prime(-3) is False
assert is_prime(0) is False
assert is_prime(1) is False
assert is_prime(2) is True
assert is_prime(3) is True
assert is_prime(4) is False
print("tests for is_prime() passed.")
```

```python
def question_1(P: List[int], Q: List[int]) -> Dict[str, Any]:
  # (a)
  R = [x for x in P if is_prime(x)] + [x for x in Q if is_prime(x)]

  # (b)
  product = P[5] * Q[11] # Index starts from 0, so the index of the nth element is n - 1

  # (c)
  Q_updated = Q.copy() # Create a copy to avoid modifying the original array (preserve input data)
  Q_updated[9] = product # Index of 9, which is the 10th element. We will replace 33 with 714

  # (d)
  complexity = "O(len(P) + len(Q))" # Each array (P and Q) is traversed once, so the time complexity is the sum of their lengths

  return {
    "R": R,
    "product": product,
    "Q_updated": Q_updated,
    "complexity": complexity,
  }

# Run Question 1
q1_result = question_1(P1, Q1)
q1_result
```


## Question 2(a) — Manual Heap Sort walkthrough
Write your manual steps in this markdown section.

Suggested format:
1. Original array
2. Build max-heap
3. Extraction 1
4. Array after MAX-HEAPIFY
5. Extraction 2
6. Array after MAX-HEAPIFY
7. Continue until sorted


```python

def max_heapify(arr: List[int], n: int, i: int) -> None:
  """Ensure the subtree rooted at index i satisfies the max-heap property."""
  largest = i
  left = 2 * i + 1
  right = 2 * i + 2

  if left < n and arr[left] > arr[largest]:
        largest = left
  if right < n and arr[right] > arr[largest]:
        largest = right

  if largest != i:
    arr[i], arr[largest] = arr[largest], arr[i]
    max_heapify(arr, n, largest)


def build_max_heap(arr: List[int]) -> None:
  """Convert arr into a max-heap in-place."""
  n = len(arr)
  for i in range(n // 2 - 1, -1, -1):
    max_heapify(arr, n, i)


def heap_sort(arr: List[int]) -> List[int]:
  """Return a sorted copy of arr using heap sort."""
  a = arr.copy()
  n = len(a)
  build_max_heap(a)
  for end in range(n - 1, 0, -1):
    a[0], a[end] = a[end], a[0]
    max_heapify(a, end, 0)
  return a

heap_sort(P2)
```


## Question 2(b) — Manual Quick Sort walkthrough
Write your manual steps in this markdown section.

Suggested format:
1. Chosen pivot
2. Partition result
3. Left and right recursive subarrays
4. Repeat until complete


```python

def partition(arr: List[int], low: int, high: int) -> int:
  """Lomuto partition scheme using the last element as pivot."""
  pivot = arr[high]
  i = low - 1
  for j in range(low, high):
    if arr[j] <= pivot:
      i += 1
      arr[i], arr[j] = arr[j], arr[i]
  arr[i + 1], arr[high] = arr[high], arr[i + 1]
  return i + 1


def quick_sort_inplace(arr: List[int], low: int, high: int) -> None:
  if low < high:
    pi = partition(arr, low, high)
    quick_sort_inplace(arr, low, pi - 1)
    quick_sort_inplace(arr, pi + 1, high)


def quick_sort(arr: List[int]) -> List[int]:
  """Return a sorted copy of arr using quick sort."""
  a = arr.copy()
  quick_sort_inplace(a, 0, len(a) - 1)
  return a

quick_sort(Q2)
```

```python
A3 = [15, 13, 9, 5, 12, 8, 7, 4, 0, 6, 2]
VALUE_TO_INSERT = 11
```


## Question 3 — Manual walkthrough
Write the transformation step by step here.

Suggested outline:
1. Append 11 at the end
2. Compare with parent
3. Swap if child > parent
4. Repeat until heap property is restored
5. Final heap


```python
def heap_insert(heap: List[int], key: int) -> Tuple[List[int], List[List[int]]]:
  """
  Insert key into a max-heap and return:
  - final heap
  - snapshots after each significant step
  """
  h = heap.copy()
  snapshots = [h.copy()]

  # Step 1: append at the end
  h.append(key)
  snapshots.append(h.copy())

  # Step 2: bubble up
  i = len(h) - 1
  while i > 0:
    parent = (i - 1) // 2
    if h[parent] >= h[i]:
        break
    h[parent], h[i] = h[i], h[parent]
    snapshots.append(h.copy())
    i = parent

  return h, snapshots

q3_final_heap, q3_steps = heap_insert(A3, VALUE_TO_INSERT)
print("Final heap:", q3_final_heap)
print("Steps:")
for idx, step in enumerate(q3_steps):
  print(f"Step {idx}: {step}")
```


## Question 3 — Complexity explanation
Draft your final explanation here.

Suggested idea:
Heap insertion may move the new key upward along a single root-to-leaf path.
The height of a heap with `n` elements is `O(log n)`, so at most `O(log n)` swaps/comparisons are needed.



# Question 4 — Hash table with linear probing

## What the question asks
Hash table size: `m = 11`  
Hash function: `h(k) = k mod 11`  
Collision handling: **linear probing**

Insert, in order:
`22, 1, 13, 11, 24, 33, 44`

You must show:
- table after each insertion
- first key that causes a collision
- number of probes per insertion
- one disadvantage of linear probing shown by this example


```python

HASH_KEYS = [22, 1, 13, 11, 24, 33, 44]
TABLE_SIZE = 11
```

```python

def linear_probe_insert(keys: List[int], m: int) -> Dict[str, Any]:
  """Insert keys into a hash table with linear probing and record all steps."""
  table = [None] * m
  states = []
  probes_per_key = {}
  first_collision_key = None

  for key in keys:
    start = key % m
    idx = start
    probes = 1

    if table[idx] is not None and first_collision_key is None:
        first_collision_key = key

    while table[idx] is not None:
        idx = (idx + 1) % m
        probes += 1

    table[idx] = key
    probes_per_key[key] = probes
    states.append((key, table.copy()))

  return {
    "states": states,
    "probes_per_key": probes_per_key,
    "first_collision_key": first_collision_key,
    "final_table": table,
  }

q4_result = linear_probe_insert(HASH_KEYS, TABLE_SIZE)
q4_result
```

```python

# Pretty-print Question 4 output
for key, state in q4_result["states"]:
  print(f"After inserting {key}: {state}")

print("\nFirst key causing a collision:", q4_result["first_collision_key"])
print("Probes per key:", q4_result["probes_per_key"])
print("Final table:", q4_result["final_table"])
```


## Question 4 — Written explanation
Write your answer for part (d) here.

# Question 5 — Minimum Spanning Tree (Kruskal and Prim)

## What the question asks
For the weighted undirected graph:

```python

vertices = ["A", "B", "C", "D", "E", "F", "G"]
edges = [
  ("A", "B", 7),
  ("A", "D", 5),
  ("B", "C", 8),
  ("B", "D", 9),
  ("B", "E", 7),
  ("C", "E", 5),
  ("D", "E", 15),
  ("D", "F", 6),
  ("E", "F", 8),
  ("E", "G", 9),
  ("F", "G", 11),
]
```


## Question 5(a) — Manual Kruskal walkthrough
Write the sorted edges and selection/rejection decisions here.


```python

class DisjointSet:
  def __init__(self, items: List[str]):
    self.parent = {item: item for item in items}
    self.rank = {item: 0 for item in items}

  def find(self, x: str) -> str:
    if self.parent[x] != x:
        self.parent[x] = self.find(self.parent[x])
    return self.parent[x]

  def union(self, a: str, b: str) -> bool:
    ra = self.find(a)
    rb = self.find(b)
    if ra == rb:
        return False

    if self.rank[ra] < self.rank[rb]:
        self.parent[ra] = rb
    elif self.rank[ra] > self.rank[rb]:
        self.parent[rb] = ra
    else:
        self.parent[rb] = ra
        self.rank[ra] += 1
    return True


def kruskal_mst(vertices: List[str], edges: List[Tuple[str, str, int]]) -> Tuple[List[Tuple[str, str, int]], int]:
  ds = DisjointSet(vertices)
  mst = []
  total_weight = 0

  for u, v, w in sorted(edges, key=lambda x: x[2]):
    if ds.union(u, v):
      mst.append((u, v, w))
      total_weight += w

  return mst, total_weight

kruskal_mst(vertices, edges)
```


## Question 5(b) — Manual Prim walkthrough
Write the step-by-step growth of the tree from vertex `A` here.


```python

import heapq

def prim_mst(vertices: List[str], edges: List[Tuple[str, str, int]], start: str) -> Tuple[List[Tuple[str, str, int]], int]:
  graph = {v: [] for v in vertices}
  for u, v, w in edges:
    graph[u].append((w, u, v))
    graph[v].append((w, v, u))

  visited = set([start])
  heap = graph[start][:]
  heapq.heapify(heap)

  mst = []
  total_weight = 0

  while heap and len(visited) < len(vertices):
    w, u, v = heapq.heappop(heap)
    if v in visited:
      continue
    visited.add(v)
    mst.append((u, v, w))
    total_weight += w

    for edge in graph[v]:
      if edge[2] not in visited:
        heapq.heappush(heap, edge)

  return mst, total_weight

prim_mst(vertices, edges, "A")
```

```python

# Combined Question 5 output
kruskal_result = kruskal_mst(vertices, edges)
prim_result = prim_mst(vertices, edges, "A")

print("Kruskal MST:", kruskal_result[0])
print("Kruskal total weight:", kruskal_result[1])

print("\nPrim MST:", prim_result[0])
print("Prim total weight:", prim_result[1])
```
